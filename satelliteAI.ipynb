```python
!pip install tensorflow scikit-learn pandas numpy matplotlib keras_tuner
import keras_tuner as kt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.model_selection import KFold
from google.colab import drive
drive.mount('/content/drive')

# Load and preprocess data
platform_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Capstone Related/platform_position.csv")
satellite_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Capstone Related/satellite_positions.csv")

platform_df["Time"] = pd.to_datetime(platform_df["Time"], format="%d-%b-%Y %H:%M:%S", errors="coerce")
satellite_df["Time"] = pd.to_datetime(satellite_df["Time"], format="%d-%b-%Y %H:%M:%S", errors="coerce")

platform_df = platform_df.dropna(subset=["Time"])
satellite_df = satellite_df.dropna(subset=["Time"])

# Merge datasets based on time
merged_df = pd.merge(satellite_df, platform_df, on="Time", suffixes=("_sat", "_plat"))

# Extract time-based feature
merged_df["MinuteOfDay"] = merged_df["Time"].dt.hour * 60 + merged_df["Time"].dt.minute

# Select features and target variable
features = ["Longitude_sat", "Latitude_sat", "MinuteOfDay"]
target = "NumConnections"
print(merged_df[features].head())
print(merged_df[target].head())
# Normalize the features
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(merged_df[features])

# Define sequence length
sequence_length = 50
# Define model builder
def build_model(hp):
    model = Sequential([
    BatchNormalization(input_shape=(sequence_length, len(features))),
    LSTM(256, activation="relu"),
    Dropout(0.2),
    Dense(1)
    ])
    optimizer = Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4]))
    model.compile(optimizer=optimizer, loss='huber', metrics=['mae'])

    return model

# Cross-validation setup
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_losses = []
cv_maes = []

for fold, (train_index, val_index) in enumerate(kf.split(scaled_features)):
    print(f"Fold {fold+1}")
    train_generator = TimeseriesGenerator(scaled_features[train_index], merged_df[target].values[train_index], length=sequence_length, batch_size=128)
    val_generator = TimeseriesGenerator(scaled_features[val_index], merged_df[target].values[val_index], length=sequence_length, batch_size=128)

    # Define tuner for each fold
    tuner = kt.Hyperband(
        build_model,
        objective='val_loss',
        max_epochs=50,
        factor=3,
        directory=f'fold_{fold+1}_tuning',
        project_name='lstm_optimization'
    )

    tuner.search(train_generator, epochs=50, validation_data=val_generator, callbacks=[EarlyStopping(monitor='val_loss', patience=10)])
    best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

    print(f"Best hyperparameters for Fold {fold+1}:")
    print(f"  LSTM units: {best_hps.get('units')}")
    print(f"  Dropout rate: {best_hps.get('dropout')}")
    print(f"  Learning rate: {best_hps.get('learning_rate')}")

    # Train best model for fold
    model_cv = tuner.hypermodel.build(best_hps)
    history_cv = model_cv.fit(train_generator, epochs=50, validation_data=val_generator, verbose=0, callbacks=[EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)])

    cv_losses.append(history_cv.history['val_loss'][-1])
    cv_maes.append(history_cv.history['val_mae'][-1])

# Print cross-validation results
print(f"Cross-Validation Loss: {np.mean(cv_losses):.4f} ± {np.std(cv_losses):.4f}")
print(f"Cross-Validation MAE: {np.mean(cv_maes):.4f} ± {np.std(cv_maes):.4f}")

# Train final model with best overall hyperparameters
best_model = tuner.hypermodel.build(best_hps)
history = best_model.fit(generator, epochs=150, verbose=1, callbacks=[EarlyStopping(monitor='loss', patience=15, restore_best_weights=True)])

# Plot training loss
plt.plot(history.history["loss"], label="Loss")
plt.plot(history.history["mae"], label="MAE")
plt.legend()
plt.title("Optimized LSTM Training Performance")
plt.show()

# Make predictions
predictions = best_model.predict(generator)

# Plot predictions vs actual values
plt.figure(figsize=(10, 5))
plt.plot(merged_df[target].values[sequence_length:], label="Actual")
plt.plot(predictions, label="Predicted", linestyle="dashed")
plt.legend()
plt.title("Optimized LSTM Predictions vs Actual")
plt.show()
```

```python
!pip install tensorflow scikit-learn pandas numpy matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error

# Mount Google Drive (if using Colab)
from google.colab import drive
drive.mount('/content/drive')

# Load and preprocess data
platform_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Capstone Related/platform_position.csv")
satellite_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Capstone Related/satellite_positions.csv")

platform_df["Time"] = pd.to_datetime(platform_df["Time"], format="%d-%b-%Y %H:%M:%S", errors="coerce")
satellite_df["Time"] = pd.to_datetime(satellite_df["Time"], format="%d-%b-%Y %H:%M:%S", errors="coerce")

platform_df = platform_df.dropna(subset=["Time"])
satellite_df = satellite_df.dropna(subset=["Time"])

# Merge datasets based on time
merged_df = pd.merge(satellite_df, platform_df, on="Time", suffixes=("_sat", "_plat"))

# Extract time-based feature
merged_df["MinuteOfDay"] = merged_df["Time"].dt.hour * 60 + merged_df["Time"].dt.minute

# Select features and target variable
features = ["Longitude_sat", "Latitude_sat", "MinuteOfDay"]
target = "NumConnections"
print(merged_df[features].head())
print(merged_df[target].head())

# Normalize the features
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(merged_df[features])

# Define sequence length
sequence_length = 50

# Create time-series sequences
generator = TimeseriesGenerator(scaled_features, merged_df[target].values, length=sequence_length, batch_size=128)

# Define GRU model
def create_gru_model():
    model = Sequential([
        BatchNormalization(input_shape=(sequence_length, len(features))),
        GRU(256, activation="relu"),
        Dropout(0.2),
        Dense(1)
    ])
    optimizer = Adam(learning_rate=1e-4, decay=1e-6)
    model.compile(optimizer=optimizer, loss="huber", metrics=["mae"])
    return model

# Cross-validation using TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)
mae_scores = []

for train_index, test_index in tscv.split(scaled_features):
    train_data, test_data = scaled_features[train_index], scaled_features[test_index]
    train_target, test_target = merged_df[target].values[train_index], merged_df[target].values[test_index]

    train_generator = TimeseriesGenerator(train_data, train_target, length=sequence_length, batch_size=128)
    test_generator = TimeseriesGenerator(test_data, test_target, length=sequence_length, batch_size=128)

    model = create_gru_model()
    history = model.fit(train_generator, epochs=150, verbose=1, callbacks=[EarlyStopping(monitor='loss', patience=15, restore_best_weights=True)])

    predictions = model.predict(test_generator)
    mae = mean_absolute_error(test_target[sequence_length:], predictions)
    mae_scores.append(mae)
    print(f"MAE for this fold: {mae}")

print(f"Mean MAE across all folds: {np.mean(mae_scores)}")

# Train the final model on the entire dataset
final_model = create_gru_model()
history = final_model.fit(generator, epochs=150, verbose=1, callbacks=[EarlyStopping(monitor='loss', patience=15, restore_best_weights=True)])

# Plot training loss
plt.plot(history.history["loss"], label="Loss")
plt.plot(history.history["mae"], label="MAE")
plt.legend()
plt.title("Training Performance")
plt.show()

# Make predictions
predictions = final_model.predict(generator)

# Plot predictions vs actual values
plt.figure(figsize=(10, 5))
plt.plot(merged_df[target].values[sequence_length:], label="Actual")
plt.plot(predictions, label="Predicted", linestyle="dashed")
plt.legend()
plt.title("GRU Predictions vs Actual")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

from google.colab import drive
drive.mount('/content/drive')

# Load and preprocess data
platform_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Capstone Related/platform_position.csv")
satellite_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Capstone Related/satellite_positions.csv")

platform_df["Time"] = pd.to_datetime(platform_df["Time"], format="%d-%b-%Y %H:%M:%S", errors="coerce")
satellite_df["Time"] = pd.to_datetime(satellite_df["Time"], format="%d-%b-%Y %H:%M:%S", errors="coerce")

platform_df = platform_df.dropna(subset=["Time"])
satellite_df = satellite_df.dropna(subset=["Time"])

# Merge datasets based on time
merged_df = pd.merge(satellite_df, platform_df, on="Time", suffixes=("_sat", "_plat"))

# Extract time-based feature
merged_df["MinuteOfDay"] = merged_df["Time"].dt.hour * 60 + merged_df["Time"].dt.minute

# Select features and target variable
#features = ["Longitude_sat", "Latitude_sat", "MinuteOfDay"]
num_lags = len(merged_df["MinuteOfDay"])//8 #45  # You can adjust this

# Create lag features for NumConnections
for lag in range(1, num_lags + 1):
    merged_df[f"NumConnections_lag_{lag}"] = merged_df["NumConnections"].shift(lag)

# Drop rows with NaN values after shifting
merged_df = merged_df.dropna()

# Re-define the feature set after adding lags
features = ["Longitude_sat", "Latitude_sat", "MinuteOfDay"] + [f"NumConnections_lag_{lag}" for lag in range(1, num_lags + 1)]
target = "NumConnections"
# Prepare the dataset (drop Time column as it's not numeric)
X = merged_df[features]
y = merged_df[target]

# Split into training and testing sets (ordered for time series)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Initialize and train a Decision Tree Regressor
dt_regressor = DecisionTreeRegressor(max_depth=120, random_state=42)
dt_regressor.fit(X_train, y_train)

# Predict on both training and testing sets
dt_train_preds = dt_regressor.predict(X_train)
dt_test_preds = dt_regressor.predict(X_test)

# Concatenate predictions to match full length of original y
dt_full_preds = np.concatenate([dt_train_preds, dt_test_preds])

# Evaluate only on test set
dt_mae = mean_absolute_error(y_test, dt_test_preds)
print(f"Decision Tree MAE: {dt_mae}")

# Plot predictions vs actual values over full range
plt.figure(figsize=(14, 6))
plt.plot(y.values, label="Actual", alpha=0.7)
plt.plot(dt_full_preds, label="Predicted", linestyle="dashed")
plt.legend()
plt.title("Decision Tree Predictions vs Actual (Full Range)")
plt.xlabel("Time Step")
plt.ylabel(target)
plt.tight_layout()
plt.show()
```
